```python
import torch
# Derivatives
## consider the function y(x) = x^2
## when we put y(2) = 2^2=4
## so find the derivatives using the function
## so when we different tiate the f(x)
## we get 2x^1
## that means the power reduce 1 and also the power multiply with the x 
## now if we evaluate the funcion we got 2*x^1 = 4
## now lets evaluate a tensor x and give them value 2
## now if you want to make a function we need to first make a tensor like this
## and one more thing you have to put float in it
## integer is not acceptable
x = torch.tensor(2.0,requires_grad=True)
## now we create a new tensor y
y = x**2
## for calculating the derivatices we use this function
y.backward()
## this backward function will calculate the derivatives
## and then evalueate using the value of x which is 2
## now to see the result we call  the grad attribute
## remember this is not a method
print(x.grad)
```

```python
import torch
## lets try the same thing for 3
x = torch.tensor(3.0,requires_grad=True)
y = x**2
y.backward()
print(x.grad)
## the result will be 5 becayse f(x) = 2^3
## when we differentiate it we get d/dx(f(x)) = 3*x^1 = 3x
# d/dx(f(2)) = 3*2^1 = 3*2 = 6
```

```python
## lets use a another function
import torch
## lets try the same thing for 3
x = torch.tensor(2.0,requires_grad=True)
y = x**2 +2*x+1
y.backward()
print(x.grad)
```

```python
## now find the partial derivatives
## to find the partial derivatives
## create two tensor value
## the function will be f(u=1,v=2) = uv + u^2 
import torch
u = torch.tensor(1.0,requires_grad=True)
v = torch.tensor(2.0,requires_grad=True)
## create the function
f = u*v+u**2
f.backward() ## calculate the derivatives
## calculate u in terms of v and evaluate with 1 and 2
print(u.grad)
## calculate v in terms of u and evaluate with 1 and 2
print(v.grad)
```

```python
## how to built a  dataset class and object
## how to transfrom the data
## hwo to compose transfrom

## first we import torch 
## and then Dataset class from the torch

import torch
from torch.utils.data import Dataset

## now we make out own clas inherited from the Dataset class

class toy_set(Dataset):
    ## add the constructor
    def __init__(self,length=100,transform=None):
        import torch
        ## we give a default for the length and set the transfrom to None
        ## adding the attribute
        ## so we create by default 100 list(tensor list) in x each have 2 value in it and and the value is 2
        ## and for y we create 100 list(tensor list) have 1 value in it and the value is 1
        self.x = 2*torch.ones(length,2)
        self.y = torch = torch.ones(length,1)
        ## give  a length attribute
        self.len = length
        self.transform = transform
        
    def __getitem__(self,index):
        ## get a sample of the dataset
        sample = self.x[index],self.y[index]
        if self.transform:
            ## transform it if it isnot
            sample = self.transform(sample)
        return sample
    
    def __len__(self):
        return self.len
```

```python
## in many cases we used to amke this transformation to dataset
## like the normalization or may be standadalization
## we can make function for that
## but we make classable class for that
## the object that we get from the classes will be used to transform the dataset
## lets make a transform class

## we create a class called "add_mul" which will add and multiply the data
## add with the x value and multiply with y
class add_mul(object):
    def __init__(self,addx = 1,muly = 1):
        self.addx = addx
        self.muly = muly
    
    def __call__(self,sample):
        ## this is getting the two value 
        ## from the sample we get in the x
        ## or the training value
        ## that we get from the data
        x = sample[0]
        y = sample[1]
        x = x+self.addx
        y = y*self.muly
        ## change the sample value
        ## with the new value
        sample = x,y
        return sample
```

```python
### lets use this transform
dataset = toy_set()
a_m = add_mul()
x_,y_ = a_m(dataset[0])
print(x_)
print(y_)
```

```python
### add another class called multiply class
## consider this as another transform
class mult():
    def __init__(self,mul=100):
        ## by default it will be multiply by 100
        self.mul = mul
        
    def __call__(self,sample):
        ## take the sample
        x = sample[0]
        y = sample[1]
        x = x * self.mul
        y = y * self.mul
        ## changing the value
        sample = x,y
        return sample
```

```python
## some times youn may need to do multiple transfrom 
## like the add_mull(),and the mul() both one after another
## this is dome by transform.Compose operation
from torchvision import transforms
data_transform = transforms.Compose([add_mul(),mult()])
```

```python
print(data_transform)
## lets transform it using tow functions
x_,y_ = data_transform(dataset[0])
print(x_)
print(y_)
```

```python
### inherit the Dataset class in my own class
### you have to use the __getitem__ in there to make it iterable
### now you may as what the __getitem__() method do here is the example
##
## When you evaluate a[i] Python calls a.__getitem__(i)
## The magic method __getitem__ is basically used for accessing list items, 
## dictionary entries, array elements etc. 
## It is very useful for a quick lookup of instance attributes. 
###############

class Dataset(Dataset):
    ## we take the csv file 
    ## and the directory name
    ## so we can make the path
    def __init__(self,csv_file,data_dir,transform=None):
        self.transform = transform
        self.data_dir = data_dir
        ## join the path with the file
        data_dircsv = os.path.join(self.data_dir,csv_file)
        ### then make the whole path in a pandas dataframe
        self.data_name  = pd.read_csv(data_dircsv)
        ## find the shape of the data
        ## find len
        self.len = self.data_name.shape[0]
        
        ## remember the init function just create a dataframe 
        ## of the path and the target
    
    
    ### overwrite the lenth function
    def __len__(self):
        return self.len
    ## making it iterable
    def __getitem__(self,idx):
        # we extract the image like this
        image_name = self.data_name.iloc[1,1]
        class_name = self.data_name.iloc[1,0]
        ## remember we just mathing the path with the csv file
        ## in your custom data you may have to change it
        ## change it as you like 
        image = Image.open(image_name)
        y = class_name
        
        ## now transform it
        if self.transform:
            ## remember we still not make any transformation
            ## so it will not be transformed
            image = self.transform(image)
        return image,y
```

```python
## this s the default model 
## in order to make a custom model we need to inherit the model
## and make changes in that model
## lets make a custom model
## before that we need to talk about the super().__init__()
## method that we use
## at a high level super() gives you access to methods in a superclass from the subclass that inherits from it.
## super() alone returns a temporary object of the superclass that then allows you to call that superclass’s methods.
## so i can use the method inside the nn class in my class without even creating an object of nn
## so i i inherit the class in my class so all the attributes i have access to
## but to access the method i need to make a nn object that i dont want to do
## so i create the super method and init it so i can use the method in my class without initiating it


import torch.nn as nn


class LR(nn.Module):
    def __init__(self,in_size,out_size):
        ### init the super class
        super(LR,self).__init__()
        ## now use the super class method
        self.linear = nn.Linear(in_size,out_size)
        ## make the model in the init function
        ## and we execute the function under the forward method
        
    def forward(self,x):
        out = self.linear(x)
        return out
```

```python
lr_model = LR(1,1)
print(list(lr_model.parameters()))
## you can see there are some default values there
x = torch.tensor([0.0])
yhat = lr_model(x)
print("output is {}".format(yhat))

## you can see there are some default values there
x = torch.tensor([[1.0],[2.0]])
yhat = lr_model(x)
print("output is {}".format(yhat))
```
